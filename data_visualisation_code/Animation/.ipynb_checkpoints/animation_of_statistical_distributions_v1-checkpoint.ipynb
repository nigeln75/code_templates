```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
import matplotlib.gridspec as gridspec
%matplotlib notebook

x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)

n = 100
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    
    x1hist.cla()
    x2hist.cla()
    x3hist.cla()
    x4hist.cla()
    
    x1hist.hist(x1[:curr], bins=10, alpha=0.25, color='red')
    x2hist.hist(x2[:curr], bins=10, alpha=0.25, color='blue')
    x3hist.hist(x3[:curr], bins=10, alpha=0.25, color='green')
    x4hist.hist(x4[:curr], bins=10, alpha=0.25, color='yellow')
    
    x1hist.set_title('Normal', size=12)
    x1hist.set_ylabel('Frequency', size=12)
    x2hist.set_title('Gamma', size=12)
    x2hist.set_ylabel('Frequency', size=12)
    x2hist.set_xlabel('Value', size=12)
    x3hist.set_title('Exponential', size=12)
    x4hist.set_title('Uniform', size=12)
    x4hist.set_xlabel('Value', size=12)

    fig.suptitle('4 Random Distributions: \n' + 'n = {}'.format(curr), size=16)
```

```python
fig = plt.figure(figsize=(9,9))
gspec = gridspec.GridSpec(2,2)

x1hist = plt.subplot(gspec[0, 0], frameon=False)
x1hist.spines['right'].set_visible(False)
x1hist.spines['top'].set_visible(False)
x1hist.yaxis.set_ticks_position('left')
x1hist.xaxis.set_ticks_position('bottom')

x2hist = plt.subplot(gspec[1, 0], frameon=False)
x2hist.spines['right'].set_visible(False)
x2hist.spines['top'].set_visible(False)
x2hist.yaxis.set_ticks_position('left')
x2hist.xaxis.set_ticks_position('bottom')

x3hist = plt.subplot(gspec[0, 1], frameon=False)
x3hist.spines['right'].set_visible(False)
x3hist.spines['top'].set_visible(False)
x3hist.yaxis.set_ticks_position('left')
x3hist.xaxis.set_ticks_position('bottom')

x4hist = plt.subplot(gspec[1, 1], frameon=False)
x4hist.spines['right'].set_visible(False)
x4hist.spines['top'].set_visible(False)
x4hist.yaxis.set_ticks_position('left')
x4hist.xaxis.set_ticks_position('bottom')

a = animation.FuncAnimation(fig, update, interval=100)
```
