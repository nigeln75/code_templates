```python
# Reading in a flat file
# A flat file is text file of table data containing records where a record is a row of fields or attributes
# Flat files may have a header to describe column names
# You will normally use Numpy or Pandas to import flat files
# Numpy will import data as a Numpy array
# Numpy is the de facto standard for storing numerical data - if file only contains numerical data!
# Numpy is also essential for SciKitLearn
import numpy as np
filename = 'mpg.csv'
data = np.loadtxt(filename, delimiter=',', skiprows = 1, usecols = [3,4])
data
```

```python
# Importing tab delimited files
# Import numpy
import numpy as np

# Assign the filename: file
file = 'digits_header.txt'

# Load the data: data
data = np.loadtxt(file, delimiter='\t', skiprows=1, usecols=[0,2])

# Print data
print(data)
```

```python
# If you have to import datasets which have different datatypes in different columns
# One column may contain strings and another floats, for example
# The function np.loadtxt() will freak at this
# There is another function, np.genfromtxt(), which can handle such structures
# If we pass dtype=None to it, it will figure out what types each column should be
data = np.genfromtxt('titanic.csv', delimiter=',', names=True, dtype=None)
# Names = true specifies that the file has a header
# Because the data are of different types, data is an object called a structured array
# Because numpy arrays have to contain elements that are all the same type, the structured array solves this by being a 1D array, where each element of the array is a row of the flat file imported
```

# Using read_csv() to reading DataFrames from multiple files

```python
# Import pandas
import pandas as pd

# Read 'Bronze.csv' into a DataFrame: bronze
bronze = pd.read_csv('Bronze.csv')

# Read 'Silver.csv' into a DataFrame: silver
silver = pd.read_csv('Silver.csv')

# Read 'Gold.csv' into a DataFrame: gold
gold = pd.read_csv('Gold.csv')

# Print the first five rows of gold
print(gold.head())
```

# Reading DataFrames from multiple files in a loop

```python
# Import pandas
import pandas as pd

# Create the list of file names: filenames
filenames = ['Gold.csv', 'Silver.csv', 'Bronze.csv']

# Create the list of three DataFrames: dataframes
dataframes = []
for filename in filenames:
    dataframes.append(pd.read_csv(filename))

# Print top 5 rows of 1st DataFrame in dataframes
print(dataframes[0].head())
```

# Combining DataFrames from multiple data files

```python
# Import pandas
import pandas as pd

# Make a copy of gold: medals
medals = gold.copy()

# Create list of new column labels: new_labels
new_labels = ['NOC', 'Country', 'Gold']

# Rename the columns of medals using new_labels
medals.columns = new_labels

# Add columns 'Silver' & 'Bronze' to medals
medals['Silver'] = silver['Total']
medals['Bronze'] = bronze['Total']

# Print the head of medals
print(medals.head())
```
